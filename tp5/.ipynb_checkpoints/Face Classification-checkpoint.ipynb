## Classification des visages
 
----

La classification des visages est un problème courant en vision par ordinateur. De Facebook aux aéroports, l'identification des individus a automatiquement de nombreux services publics possibles. Dans ce cahier, nous allons utiliser cette application comme test-case pour les réseaux neuronaux convolutionnels 2D (CNN).

On sait depuis longtemps que la classification des visages n’est pas un problème *difficile*, du moins par rapport à la reconnaissance d’objet en cas général et à la classification des images sans contrainte. Identifier les régions de visage dans une image demande un peu de travail, mais la structure géométrique du visage est très régulière et peut être très bien modélisée simplement par analyse en composantes principales (ACP). En se projetant sur un espace de faible dimension, il devient pratique d'entraîner des classificateurs via une régression logistique ou des machines à vecteurs de support (SVM). Souvent, cela résout assez bien le problème.

Utiliser CNN pour cette tâche est un peu excessif, mais cela nous pose un problème sur lequel nous pouvons facilement travailler avec des données d’entraînement limitées, une puissance d’entraînement limitée (ordinateurs portables en <10 minutes), tout en obtenant un résultat raisonnable.
Nous allons travailler avec un jeu de données appelé `Labelled Faces in the Wild` (LFW). [Vous pouvez trouver la description complète du jeu de données ici](http://vis-www.cs.umass.edu/lfw/). En outre, vous devrez télécharger le jeu de données (~ 100 Mo) et le stocker sur votre disque dur. La version du jeu de données que nous souhaitons utiliser est la version **Deep Funneled** qui contient des images pré-alignées et centrées des faces. Cela nous aide à éviter beaucoup de pré-traitements compliqués.

* [Download Link for **Deep Funneled** LFW Dataset.](http://vis-www.cs.umass.edu/lfw/lfw-deepfunneled.tgz)

Nous avons préparé une fonction qui s'occupe déjà du chargement et du formatage des données. Nous allons travailler avec un petit sous-ensemble du jeu de données, celui-ci ne contenant que des individus pour lesquels il existe plus de ~ 50 images de visage. Cela réduit la quantité de données à utiliser.

De plus, les images originales sont de taille $ 250 \times 250 $ pixels. C'est un peu plus grand que ce que nous voulons utiliser sur nos ordinateurs portables, nous réduisons donc les images à une taille de $125  x 125 $ pixels. La fonction de chargement applique également un brassage du jeu de données. Nous définissons donc l'état du générateur de nombres aléatoires (kwarg `random_state`) afin que nous nous approchions tous des mêmes résultats pour cette démonstration.

```python
#--- Load our LFW Data ---#
(X, y, names) = load_lfw_dataset(min_num_faces=49, scale=0.5, data_home='./mldata', random_state=42)
print('Dataset Dimensions: ', X.shape)
```

Jetons un coup d'oeil sur le jeu de données d'une manière abrégée. Ceci est juste pour le plaisir et pour avoir une idée des données. Nous allons faire un portrait de chaque individu dans le jeu de données par une simple moyenne de leurs images pour afficher la moyenne résultante.

```python
#--- View People in Dataset ---#
num_people = len(names)

L = int(np.sqrt(num_people)) + 1

plt.figure(figsize=(15,15))

for i in range(0,num_people):
    plt.subplot(L,L,i+1)
    plt.imshow(np.mean(X[y==i, :, :, :], axis=0))
    plt.title(names[i], fontsize=16)

plt.tight_layout()
```

Une facette importante de la ML efficace consiste à se concentrer sur les caractéristiques prédictives connues. Dans ce cas, nous souhaitons avoir un modèle axé sur la classification des *visages* et non de l'arrière-plan. Nous pouvons donc réduire la complexité de notre modèle en effectuant un recadrage au centre des images de visage.

```python
#--- Crop to Faces ---#
# Get crop dimensions
vert_dim  = X.shape[1]
horz_dim  = X.shape[2] 
horz_crop = (int(0.2*horz_dim), int(0.8*horz_dim))
vert_crop = (int(0.08*vert_dim), int(0.84*vert_dim))   
# Make crop
X = X[:, vert_crop[0]:vert_crop[1], horz_crop[0]:horz_crop[1], :]
print('Dataset Dimensions: ', X.shape)
# Save this new dimension
img_dim = (X.shape[1], X.shape[2], 3)

#--- Display ---#
plt.figure(figsize=(15,15))

for i in range(0,num_people):
    plt.subplot(L,L,i+1)
    plt.imshow(np.mean(X[y==i, :, :, :], axis=0))
    plt.title(names[i], fontsize=16)

plt.tight_layout()
```

Nous allons maintenant procéder à une division train-test et mettre de côté quelques données de test afin de comparer nos modèles.

```python
from sklearn.model_selection import train_test_split

X, X_test, y, y_test = train_test_split(X, y, random_state=42, test_size=0.1)

print('Train Dataset Shape: ', X.shape)
print('Test Dataset Shape: ', X_test.shape)
```

Voyons maintenant ce que nous pouvons savoir de cet ensemble de données de la manière la plus simple possible. Ceci est juste pour avoir une idée des principaux axes qui séparent les images les unes des autres. Nous pouvons le faire en regardant les principales composantes (PC) de l’ensemble de données (par exemple, la PCA). Pour ce faire, nous devons d’abord aplatir le jeu de données.

```python
#--- Take a look at the PCA ---#
def flatten(X):
    XFlat = np.empty((X.shape[0], X.shape[1]*X.shape[2]*X.shape[3]), dtype=float)
    for i in range(X.shape[0]):
        XFlat[i,:] = np.squeeze(X[i, :, :, :]).ravel() 
    
    return XFlat
    

XFlat = flatten(X)
XFlat_test = flatten(X_test)

print('Flattened Dimensions: ', XFlat.shape)           
```

Nous pouvons maintenant utiliser Scikit-Learn pour trouver les PC.

```python
from sklearn.decomposition import PCA

#--- Calculate PCA Transform ---#
pca = PCA(n_components=100)
pca.fit(XFlat)
# Find projection of dataset onto PCs
XPCA = pca.transform(XFlat)
# Find values of the PCs themselves
PCS = pca.components_
```

Voyons ce que sont les PC. Étant donné que ces PC sont générés à partir d'un SVD, il s'agit essentiellement de vecteurs à base orthogonale qui décrivent le mieux la variation dans l'ensemble de données. De plus, scikit-learn donne les composants de manière ordonnée, par ex. Le premier PC est le vecteur qui décrit la direction maximale de variation dans le jeu de données. Ci-dessous, nous traçons les 25 premiers PC.

```python
plt.figure(figsize=(15,15))

for i in range(25):
    plt.subplot(5,5,i+1)
    comp = PCS[i,:]
    comp = (comp - np.min(comp)) / (np.max(comp) - np.min(comp))
    plt.imshow(np.reshape(comp,(95,75,3)))
    plt.title('PC %s' % str(i+1), fontsize=16)

plt.tight_layout()
```

Maintenant, regarder ces ordinateurs ne nous en dit pas beaucoup. Cependant, ce que nous pouvons faire, c'est examiner la projection du jeu de données de faible temps en mouvement sur ces principaux composants principaux. Nous pouvons ensuite classer l'ensemble de données en fonction de l'ordre de ces composants. Certaines images de visage peuvent avoir de grandes valeurs positives ou négatives lorsqu'elles sont projetées sur un PC particulier. En regardant les scores les plus grands et les plus petits de l’ensemble de données par rapport au PC, nous en apprenons un peu plus sur la séparation le long du PC.

Ci-dessous, nous présentons les 10 PC les plus importants, ainsi que les images de jeu de données d'évaluation les plus grandes et les plus petites.

```python
plt.figure(figsize=(15,30))
top_show = 3
pc_show  = 10
pltcnt = 1
for pc in range(pc_show):
    top_scoring = np.argsort(XPCA[:, pc])[-top_show:]
    bot_scoring = np.argsort(XPCA[:, pc])[top_show:]

    # Show PC
    plt.subplot(pc_show, (2*top_show)+1, pltcnt)  
    comp = PCS[pc,:]
    comp = (comp - np.min(comp)) / (np.max(comp) - np.min(comp))
    plt.imshow(np.reshape(comp,img_dim))
    plt.title('PC %s' % str(pc+1), fontsize=16)

    pltcnt+=1
    # Show hi-scoring
    for i in range(top_show):
        plt.subplot(pc_show, (2*top_show)+1, pltcnt)    
        plt.imshow(X[top_scoring[i], :, :, :])
        plt.title('Hi Score # %s' % str(i+1), fontsize=16)
        pltcnt+=1
        
    for i in range(top_show):
        plt.subplot(pc_show, (2*top_show)+1, pltcnt)    
        plt.imshow(X[bot_scoring[i], :, :, :])
        plt.title('Lo Score # %s' % str(i+1), fontsize=16)
        pltcnt+=1        
        
plt.tight_layout()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

#--- Fit the Perceptron ---#
mlr = LogisticRegression(solver='sag', max_iter=1000, multi_class='multinomial')
mlr.fit(XPCA, y)

#--- Transform and predict test data ---#
ypred = mlr.predict(pca.transform(XFlat_test))

#--- Show results on test data ---#
print(classification_report(y_test, ypred, target_names=names))

#--- Calculate Confusion Matrix ---#
cmat = confusion_matrix(y_test, ypred).astype(np.float)
# Normalization...
for i in range(cmat.shape[0]):
    cmat[i,:] = cmat[i,:] / np.sum(cmat[i,:])
# Display...
plt.figure(figsize=(10,10))    
plt.imshow(cmat, vmin=0, vmax=1, cmap='Blues') 
plt.yticks(range(len(names)),names)
plt.xticks(range(len(names)),names,rotation='vertical')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.colorbar()
plt.tight_layout()
```

Ainsi, comme nous pouvons le constater, nous sommes en mesure d’obtenir un niveau de performance de classification assez adéquat grâce à une régression logistique multinomiale des 100 principales composantes. C'est une approche simple qui nous donne un classificateur efficace. Maintenant la question reste, pouvons-nous faire beaucoup mieux? Il existe certainement d'autres approches que nous pourrions utiliser (CARTS, Boosting, etc.) pour la classification / régression. Cependant, profitons de l’occasion pour appliquer les CNN à cette tâche de classification d’images.

## CNNs 1
---

Voyons maintenant comment utiliser un CNN pour cette même tâche. Dans ce cahier, nous allons nous concentrer sur une structure particulière et la laisser ensuite comme exercice pour tenter de la renforcer. Notre structure initiale sera:

* **[Input] -> [Conv -> ReLu -> Conv -> ReLu -> MaxPool]x2 -> Dense -> ReLu -> Dropout -> Labels **

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Conv2D, MaxPooling2D, Dropout, Flatten
from keras.utils import to_categorical
from keras.optimizers import SGD

input_dim = (X.shape[1], X.shape[2], X.shape[3])
#--- Network Structure ---#
faceNet = Sequential()

# First Convolution Block
faceNet.add(Conv2D(32, 
                   kernel_size=(3,3), 
                   data_format='channels_last', 
                   input_shape=input_dim, 
                   activation='relu'))
faceNet.add(Conv2D(32, 
                   kernel_size=(3,3), 
                   activation='relu'))
faceNet.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))

# Second Convolution Block
faceNet.add(Conv2D(32, 
                   kernel_size=(3,3), 
                   activation='relu'))
faceNet.add(Conv2D(32, 
                   kernel_size=(3,3), 
                   activation='relu'))
faceNet.add(MaxPooling2D(pool_size=(2,2), strides=(2,2)))

# To Fully Connected Layer
faceNet.add(Flatten())
faceNet.add(Dense(128, activation='relu'))
faceNet.add(Dropout(0.2))

# Final Labels
faceNet.add(Dense(num_people, activation='softmax'))

#--- Training ---#
# Set Parameters
faceNet.compile(optimizer='adam', 
                loss='categorical_crossentropy', 
                metrics=['accuracy'])

# Get class weights
class_counts = np.histogram(y, bins=num_people)[0]
class_weight = class_counts / np.max(class_counts)

# Run Training
train_history = faceNet.fit(X, to_categorical(y), validation_split=0.05, epochs=10, verbose=1, batch_size=32, class_weight=class_weight)

# Save model parameters
faceNet.save('faceNet.h5')
```

```python
# ypred = faceNet.predict(X_test)
ypred = faceNet.predict_classes(X_test)
```

```python
#--- Show results on test data ---#
print(classification_report(y_test, ypred, target_names=names))

#--- Calculate Confusion Matrix ---#
cmat = confusion_matrix(y_test, ypred).astype(np.float)
# Normalization...
for i in range(cmat.shape[0]):
    cmat[i,:] = cmat[i,:] / np.sum(cmat[i,:])
# Display...
plt.figure(figsize=(10,10))    
plt.imshow(cmat, vmin=0, vmax=1, cmap='Blues') 
plt.yticks(range(len(names)),names)
plt.xticks(range(len(names)),names,rotation='vertical')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.colorbar()
plt.tight_layout()
```

Voyons maintenant ce que nous avons manqué. Il est souvent éclairant de comprendre les types de caractéristiques communes aux échantillons erronés. Dans ce cas, nous ne devrions pas affiner notre approche car nous examinons les erreurs sur l'ensemble *test*. Cependant, sur le plan académique, nous pouvons voir que beaucoup de ces images présentent des caractéristiques atypiques. Par exemple, une personne portant un vêtement peu commun, une partie de son visage obscurcie, un éclairage ou des angles inhabituels ou même des données mal étiquetées / alignées.

```python
test_errors = np.where(np.not_equal(y_test, ypred))[0]
L = int(len(test_errors) / 5) + 1

plt.figure(figsize=(15,15))
pltidx = 1

def add_subplot_axes(ax,rect,axisbg='w'):
    fig = plt.gcf()
    box = ax.get_position()
    width = box.width
    height = box.height
    inax_position  = ax.transAxes.transform(rect[0:2])
    transFigure = fig.transFigure.inverted()
    infig_position = transFigure.transform(inax_position)    
    x = infig_position[0]
    y = infig_position[1]
    width *= rect[2]
    height *= rect[3]  # <= Typo was here
    subax = fig.add_axes([x,y,width,height],axisbg=axisbg)
    x_labelsize = subax.get_xticklabels()[0].get_size()
    y_labelsize = subax.get_yticklabels()[0].get_size()
    x_labelsize *= rect[2]**0.5
    y_labelsize *= rect[3]**0.5
    subax.xaxis.set_tick_params(labelsize=x_labelsize)
    subax.yaxis.set_tick_params(labelsize=y_labelsize)
    return subax

for i in range(len(test_errors)):
    test_id = test_errors[i]
    im = np.empty((1, X_test.shape[1], X_test.shape[2], X_test.shape[3]), dtype=float)
    im[0,:,:,:] = X_test[test_id, :, :, :]
    
    ax = plt.subplot(L, 5, pltidx)
    plt.imshow(X_test[test_id, :, :, :])
    plt.title(names[ypred[test_id]])
    sax = add_subplot_axes(ax, [0.2, 0.0, 0.7, 0.2])
    sax.bar(range(num_people),faceNet.predict(im).ravel(), color='b', edgecolor='w')
    sax.axis('off')
    
    pltidx += 1
    

```
