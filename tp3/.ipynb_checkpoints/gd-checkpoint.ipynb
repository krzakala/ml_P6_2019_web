# Descente de gradients a gogo !

Le but de ce cahier est de gagner l’intuition de diverses méthodes de descente de gradient en visualisant et en appliquant ces méthodes à des surfaces simples en deux dimensions. Les méthodes étudiées incluent la descente de gradient ordinaire, la descente de gradient avec impulsion, NAG, ADAM et RMSProp.


## Overview

Dans ce cahier, nous allons visualiser les différentes méthodes de descente de gradient effectuées à l'aide de surfaces simples. Dès le départ, nous soulignons que la descente de gradient sur les surfaces est différente de la descente de gradient sur une fonction de perte en Machine Learning (ML). La raison en est que dans ML, nous voulons non seulement trouver les bons minima, mais également les bons minima qui se généralisent bien aux nouvelles données. Malgré cette différence cruciale, nous pouvons toujours construire une intuition sur les méthodes de descente sur gradient en les appliquant à des surfaces simples (voir les articles de blog associés [ici] (http://ruder.io/optimizing-gradient-descent/) et [ici] (http://tiao.io/notes/visualizing-and-animating-optimization-algorithms-with-matplotlib/)).

## Gradient descent avec et sans momentum!

Dans ce cahier, nous allons visualiser divers algorithmes de descente de gradient utilisés dans l’apprentissage automatique. Nous serons particulièrement intéressés à essayer de comprendre comment divers hyperparamètres - en particulier le taux d’apprentissage - affectent notre performance. Ici, nous nous limitons principalement à regarder la performance en l'absence de bruit. Cependant, nous encourageons le lecteur à jouer avec la force du bruit ci-dessous et à voir les différences qui introduisent une stochasticité.

Tout au long, nous notons les paramètres par $ \theta $ et la fonction d’énergie que nous essayons de minimiser par $ E (\theta) $.

<b>Descente de gradient</b>

Nous commençons par considérer une méthode simple de descente sur gradient. Dans cette méthode,
nous allons faire des pas en direction du gradient local. Étant donné certains paramètres $ \theta $, nous ajustons les paramètres à chaque itération afin que

$$\theta_{t+1}= \theta_t - \eta_t \nabla_\theta E(\theta),$$

où nous avons introduit le taux d’apprentissage $ \eta_t $ qui contrôle l’ampleur de notre démarche. En général, l'algorithme est extrêmement sensible au choix de $ \eta_t $. Si $ \eta_t $ est trop grand, alors on peut énormément osciller autour des minima et rater une structure importante à petite échelle. Ce problème est amplifié si nos calculs de gradients sont bruyants et inexacts (comme c'est souvent le cas dans les applications d'apprentissage automatique). Si $ \eta_t $ est trop petit, la procédure d'apprentissage / minimisation devient extrêmement lente.

<b>Descente de gradient avec momentum</b>

Un problème avec la descente de gradient est qu’il n’a aucun souvenir de la "balle qui roule en bas de la colline". Cela peut poser problème lorsqu'il existe de nombreux minima peu profonds dans notre paysage. Si nous faisons une analogie avec une balle qui roule sur une colline, le manque de mémoire équivaut à ne pas avoir d’inertie ni de momentum (c’est-à-dire une dynamique complètement suramortie). Sans élan, la balle n'a pas d'énergie cinétique et ne peut pas sortir des minima peu profonds.

L'élan devient particulièrement important lorsque nous commençons à penser à la descente de gradient stochastique avec des estimations stochastiques bruyantes du gradient. Dans ce cas, nous devrions nous rappeler d'où nous venons et ne pas réagir de manière drastique à chaque nouvelle mise à jour.

nspiré par cela, nous pouvons ajouter un terme de mémoire ou de "momentum" en anglais (inertie) au terme de descente de gradient stochastique ci-dessus:

$$
v_{t}=\gamma v_{t-1}+\eta_{t}\nabla_\theta E(\theta_t),\\
\theta_{t+1}= \theta_t -v_{t},
$$

avec $ 0 \le \gamma <1 $ appelé le paramètre momentum. Lorsque $ \gamma = 0 $, cela revient à une descente de gradient ordinaire et une augmentation de $ \ gamma $ augmente la contribution inertielle au gradient. D'après les équations ci-dessus, nous pouvons voir que la durée de vie typique du gradient est donnée par $ (1- \gamma) ^ {- 1} $. Pour $ \gamma = 0 $ comme dans la descente de gradient, la durée de vie n’est qu’une étape. Pour $ \gamma = 0.9 $, nous nous souvenons généralement d'un gradient pour dix étapes. Nous appellerons cette descente de gradient avec élan classique ou CM pour faire court.

Une dernière variante largement utilisée de descente de gradient avec impulsion est appelée gradient accéléré de Nesterov (NAG). Dans NAG, plutôt que de calculer le gradient à la position actuelle, on calcule le gradient à l’élan de position qui nous mènera au moment $ t + 1 $, à savoir, $ \theta_t - \gamma v_ {t-1} $. Ainsi, la mise à jour devient
$$
v_{t}=\gamma v_{t-1}+\eta_{t}\nabla_\theta E(\theta_t-\gamma v_{t-1})\\
\theta_{t+1}= \theta_t -v_{t}
$$

```python
#This writes a simple gradient descent, gradient descent+ momentum,
#nesterov. 

#Mean-gradient based methods
def gd(grad, init, n_epochs=1000, eta=10**-4, noise_strength=0):
    #This is a simple optimizer
    params=np.array(init)
    param_traj=np.zeros([n_epochs+1,2])
    param_traj[0,]=init
    v=0;
    for j in range(n_epochs):
        noise=noise_strength*np.random.randn(params.size)
        v=eta*(np.array(grad(params))+noise)
        params=params-v
        param_traj[j+1,]=params
    return param_traj


def gd_with_mom(grad, init, n_epochs=5000, eta=10**-4, gamma=0.9,noise_strength=0):
    params=np.array(init)
    param_traj=np.zeros([n_epochs+1,2])
    param_traj[0,]=init
    v=0
    for j in range(n_epochs):
        noise=noise_strength*np.random.randn(params.size)
        v=gamma*v+eta*(np.array(grad(params))+noise)
        params=params-v
        param_traj[j+1,]=params
    return param_traj

def NAG(grad, init, n_epochs=5000, eta=10**-4, gamma=0.9,noise_strength=0):
    params=np.array(init)
    param_traj=np.zeros([n_epochs+1,2])
    param_traj[0,]=init
    v=0
    for j in range(n_epochs):
        noise=noise_strength*np.random.randn(params.size)
        params_nesterov=params-gamma*v
        v=gamma*v+eta*(np.array(grad(params_nesterov))+noise)
        params=params-v
        param_traj[j+1,]=params
    return param_traj
```


## Descente de gradient qui utilise le second moment

Dans la descente à gradient stochastique, avec et sans élan, nous devons encore spécifier un calendrier pour ajuster les vitesses d'apprentissage $ \ eta_t $ en fonction du temps. Comme discuté dans Sec. IV dans le contexte de la méthode de Newton, cela présente un certain nombre de dilemmes. Le taux d'apprentissage est limité par la direction la plus raide qui peut changer en fonction de l'endroit où vous vous trouvez dans le paysage. Pour contourner ce problème, idéalement, notre algorithme prendrait de grandes étapes dans des directions plates et peu profondes et de petites étapes dans des directions raides et étroites. Les méthodes de second ordre réalisent cela en calculant ou en approximant le hessien et en normalisant le taux d'apprentissage en fonction de la courbure. Cependant, cela coûte très cher en calcul pour des modèles extrêmement volumineux. Idéalement, nous aimerions pouvoir modifier de manière adaptative la taille de nos pas pour s’adapter au paysage sans payer le prix de calcul raide du calcul ou de l’approximation du nombre de Hesses.

Récemment, un certain nombre de méthodes ont été introduites pour y parvenir en suivant non seulement le gradient mais également le deuxième moment du gradient. Ces méthodes incluent AdaGrad, AdaDelta, RMS-Prop et ADAM. Ici, nous discutons le dernier de ces deux en tant que représentants de cette classe d'algorithmes.


Dans RMS prop (Root-Mean-Square propagation), en plus de garder une moyenne courante du premier moment de la pente, nous gardons également une trace du deuxième moment par le biais d'une moyenne mobile. La règle de mise à jour pour RMS prop est donnée par
$$
\mathbf{g}_t = \nabla_\theta E(\boldsymbol{\theta}) \\
\mathbf{s}_t =\beta \mathbf{s}_{t-1} +(1-\beta)\mathbf{g}_t^2 \nonumber \\
\boldsymbol{\theta}_{t+1}=\boldsymbol{\theta}_t + \eta_t { \mathbf{g}_t \over \sqrt{\mathbf{s}_t +\epsilon}}, \nonumber \\
$$
où $\beta$ contrôle le temps moyen du deuxième moment et est généralement pris pour environ $ \beta = 0.9 $, $ \eta_t $ est un taux d'apprentissage généralement choisi pour être $ 10 ^ {- 3} $, et $ \epsilon \sim 10 ^ {-8} $ est une petite constante de régularisation destinée à éviter les divergences. Il ressort clairement de cette formule que le taux d’apprentissage est réduit dans les directions où la norme du gradient est toujours élevée. Cela accélère grandement la convergence en nous permettant d’utiliser un taux d’apprentissage plus élevé pour les directions à plat.

Un algorithme associé est l'optimiseur ADAM. Dans ADAM, nous maintenons une moyenne courante du premier et du second moment du gradient et utilisons ces informations pour modifier de manière adaptative le taux d'apprentissage pour différents paramètres. En plus de conserver une moyenne courante des premier et second moments du gradient, ADAM effectue une correction de biais supplémentaire pour tenir compte du fait que nous estimons les deux premiers moments du gradient à l’aide d’une moyenne courante (indiquée par les chapeaux dans le diagramme). mettre à jour la règle ci-dessous). La règle de mise à jour pour ADAM est donnée par (où multiplication et division sont considérées comme des opérations élémentaires)
$$
\mathbf{g}_t = \nabla_\theta E(\boldsymbol{\theta}) \\
\mathbf{m}_t = \beta_1 \mathbf{m}_{t-1} + (1-\beta_1) \mathbf{g}_t \nonumber \\
\mathbf{s}_t =\beta_2 \mathbf{s}_{t-1} +(1-\beta_2)\mathbf{g}_t^2 \nonumber \\
\hat{\mathbf{m}}_t={\mathbf{m}_t \over 1-\beta_1} \nonumber \\
\hat{\mathbf{s}}_t ={\mathbf{s}_t \over1-\beta_2} \nonumber \\
\boldsymbol{\theta}_{t+1}=\boldsymbol{\theta}_t + \eta_t { \hat{\mathbf{m}}_t \over \sqrt{\hat{\mathbf{s}}_t +\epsilon}}, \nonumber 
$$
où $ \beta_1 $ et $ \beta_2 $ définissent la durée de vie de la mémoire du premier et du deuxième moment. Ils sont généralement pris à 0,9 $ et 0,99 $ respectivement, et $ \eta $ et $ \epsilon $ sont identiques à RMSprop.




```python
################################################################################
# Methods that exploit first and second moments of gradient: RMS-PROP and ADAMS
################################################################################

def rms_prop(grad, init, n_epochs=5000, eta=10**-3, beta=0.9,epsilon=10**-8,noise_strength=0):
    params=np.array(init)
    param_traj=np.zeros([n_epochs+1,2])
    param_traj[0,]=init#Import relevant packages
    grad_sq=0;
    for j in range(n_epochs):
        noise=noise_strength*np.random.randn(params.size)
        g=np.array(grad(params))+noise
        grad_sq=beta*grad_sq+(1-beta)*g*g
        v=eta*np.divide(g,np.sqrt(grad_sq+epsilon))
        params= params-v
        param_traj[j+1,]=params
    return param_traj
                        
                        
def adams(grad, init, n_epochs=5000, eta=10**-4, gamma=0.9, beta=0.99,epsilon=10**-8,noise_strength=0):
    params=np.array(init)
    param_traj=np.zeros([n_epochs+1,2])
    param_traj[0,]=init
    v=0;
    grad_sq=0;
    for j in range(n_epochs):
        noise=noise_strength*np.random.randn(params.size)
        g=np.array(grad(params))+noise
        v=gamma*v+(1-gamma)*g
        grad_sq=beta*grad_sq+(1-beta)*g*g
        v_hat=v/(1-gamma)
        grad_sq_hat=grad_sq/(1-beta)
        params=params-eta*np.divide(v_hat,np.sqrt(grad_sq_hat+epsilon))
        param_traj[j+1,]=params
    return param_traj
```
